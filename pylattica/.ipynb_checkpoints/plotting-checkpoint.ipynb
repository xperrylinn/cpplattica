# Python versus C++

```python
file_path_to_pylattica_serial_times = "./gol_pylattica_serial.csv"
file_path_to_cpplattica_serial_times = "./cpplattica_serial_times.csv"

pylattica_serial_df = pd.read_csv(file_path_to_pylattica_serial_times)
cpplattica_serial_df = pd.read_csv(file_path_to_cpplattica_serial_times)

# Apply log10 to the time column
pylattica_serial_df["log[time (seconds)]"] = pylattica_serial_df["time (seconds)"].apply(np.log10)
cpplattica_serial_df["log[time (seconds)]"] = cpplattica_serial_df["time (seconds)"].apply(np.log10)

plt.figure(figsize=(10, 6))

# Scatter plot with curve lines for Pylattica
plt.scatter(pylattica_serial_df["size"], pylattica_serial_df["log[time (seconds)]"], label="pylattica", color='blue', marker='o')
plt.plot(pylattica_serial_df["size"], pylattica_serial_df["log[time (seconds)]"], color='blue', linestyle='-')

# Scatter plot with curve lines for CPPLattica
plt.scatter(cpplattica_serial_df["size"], cpplattica_serial_df["log[time (seconds)]"], label="cpplattica", color='red', marker='^')
plt.plot(cpplattica_serial_df["size"], cpplattica_serial_df["log[time (seconds)]"], color='red', linestyle='-')

plt.xlabel("size")
plt.ylabel("log[time (seconds)]")
plt.title("Pylattica vs. Cpplatica\nSerial Implementation")

# Move the legend to the right outside the grid
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))

plt.grid(True)
output_file = "pylattica_versus_cpplattica_serial.png"
plt.savefig(output_file, bbox_inches="tight")
plt.show()
```

```python
file_path = "./sim_structure_nprocs_1_rank_0_size_100_nsteps_10.json"
json_file = open(file_path, 'r')
periodic_structure_dict = json.load(json_file)
```

```python
structure = PeriodicStructure.from_dict(periodic_structure_dict)
```

```python
initial_state = SimulationState.from_struct(structure)
```

```python
dead = 0
alive = 1
state = "STATE"

state_choices = [dead, alive]

for site_id in structure.site_ids:
    initial_state.set_site_state(site_id, {
        state: random.choice(state_choices)
    })
```

```python
class GoLCellArtist(CellArtist):
    
    def get_color_from_cell_state(self, cell_state):
        if cell_state[state] == dead:
            return (0,100,150)
        else:
            return (255, 255, 150)
    
    def get_cell_legend_label(self, cell_state):
        if cell_state[state] == dead:
            return "Dead"
        else:
            return "Alive"
```

```python
from pylattica.core import BasicController, SimulationState, PeriodicStructure
from pylattica.structures.square_grid import MooreNbHoodBuilder
from pylattica.discrete.state_constants import DISCRETE_OCCUPANCY



def process_variant_string(v_string):
    halves = v_string.split("/")
    born = [int(el) for el in list(halves[0][1:])]
    survive = [int(el) for el in list(halves[1][1:])]
    return born, survive
    

class GameOfLifeController(BasicController):
    variant = None

    def __init__(self, structure: PeriodicStructure, variant="B3/S23"):
        if self.variant is None:
            self.variant = variant
        self.born, self.survive = process_variant_string(self.variant)
        self.structure = structure

    def pre_run(self, _):
        self.neighborhood = MooreNbHoodBuilder().get(self.structure)

    def get_state_update(self, site_id, curr_state: SimulationState):
        alive_neighbor_count = 0
        dead_neighbor_count = 0

        neighbor_site_ids = self.neighborhood.neighbors_of(site_id)

        for nb_id in neighbor_site_ids:
            neighbor_state = curr_state.get_site_state(nb_id)[DISCRETE_OCCUPANCY]
            if neighbor_state == 1:
                alive_neighbor_count += 1
            else:
                dead_neighbor_count += 1

        current_state = curr_state.get_site_state(site_id)[DISCRETE_OCCUPANCY]

        if current_state == 1 and alive_neighbor_count in self.survive:
            new_state = 1
        elif current_state == 0 and alive_neighbor_count in self.born:
            new_state = 1
        else:
            new_state = 0

        updates = {DISCRETE_OCCUPANCY: new_state}
        return updates
```

```python
controller = GameOfLifeController(structure)
runner = SynchronousRunner(parallel=False)
result = runner.run(initial_state, controller, 100)
```

```python
grid_artist.jupyter_show(result.get_step(10), cell_size=15)
grid_artist.jupyter_show(result.get_step(11), cell_size=15)
grid_artist.jupyter_show(result.get_step(12), cell_size=15)
```
